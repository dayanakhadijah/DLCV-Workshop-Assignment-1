```python
import cv2
import matplotlib.pyplot as plt
#coins image 1
coins = cv2.imread(path+"COINS.jpg")
coins_gray = cv2.cvtColor(coins, cv2.COLOR_BGR2GRAY)

coins_blur = cv2.GaussianBlur(coins_gray, (11,11), 0)

coins_canny = cv2.Canny(coins_blur, 30, 150, 3)

coins_dilate = cv2.dilate(coins_canny, (1,1), iterations=0)

(cnt, hierarchy) = cv2.findContours(coins_dilate.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
coins_rgb = cv2.cvtColor(coins, cv2.COLOR_BGR2RGB)
cv2.drawContours(coins_rgb, cnt, -1, (255, 0, 0), 2)

plt.figure(figsize=[18,5])
plt.subplot(131); plt.imshow(coins_blur, cmap='gray'); plt.title("Blurred");
plt.subplot(132); plt.imshow(coins_dilate, cmap='gray'); plt.title("Canny Dilated");
plt.subplot(133); plt.imshow(coins_rgb); plt.title("Contoured");

print("Coins in Image 1 = ", len(cnt))

#coins image 2
coins2 = cv2.imread(path+"COINS2.png")
coins2_gray = cv2.cvtColor(coins2, cv2.COLOR_BGR2GRAY)

coins2_blur = cv2.GaussianBlur(coins2_gray, (13,13), 0)

coins2_canny = cv2.Canny(coins2_blur, 30, 150, 3)

coins2_dilate = cv2.dilate(coins2_canny, (1,1), iterations=3)

(cnt2, hierarchy) = cv2.findContours(coins2_dilate.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
coins2_rgb = cv2.cvtColor(coins2, cv2.COLOR_BGR2RGB)
cv2.drawContours(coins2_rgb, cnt2, -1, (255, 0, 0), 2)

plt.figure(figsize=[18,5])
plt.subplot(131); plt.imshow(coins2_blur, cmap='gray'); plt.title("Blurred");
plt.subplot(132); plt.imshow(coins2_dilate, cmap='gray'); plt.title("Canny Dilated");
plt.subplot(133); plt.imshow(coins2_rgb); plt.title("Contoured");
print("Coins in Image 2 = ", len(cnt2))
```
