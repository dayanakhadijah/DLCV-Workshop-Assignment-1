```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image 
%matplotlib inline
from IPython.display import Image
```

```python
path = "/content/gdrive/MyDrive/DLCV_ASSIGNMENT 1/FACE PICS/"
```

```python
face_cascade = cv2.CascadeClassifier(cv2.samples.findFile(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'))
```

```python
#face1 image convert to rgb
face1_bgr= cv2.imread(path+"Face1.jpg")
face1 = cv2.cvtColor(face1_bgr, cv2.COLOR_BGR2RGB)

#crop image
cropped_face1 = face1[0:400, 150:500]

#transparent overlay for bounding box
bbox_array = np.zeros([480,640,4], dtype=np.uint8)

#grayscale image for face detection
face1_gray = cv2.cvtColor(cropped_face1, cv2.COLOR_RGB2GRAY)

#get face region coordinates
face1_data = face_cascade.detectMultiScale(face1_gray)

#Draw rectangle around the faces which is our region of interest(ROI)
for (x,y,w,h) in face1_data:
  bbox_array = cv2.rectangle(bbox_array,(x,y),(x+w,y+h),(255,0,0),2)
  roi = cropped_face1[y:y+h, x:x+w]
  roi = cv2.GaussianBlur(roi, (21,21), 5)
  cropped_face1[y:y+roi.shape[0], x:x+roi.shape[1]] = roi

plt.figure(figsize=[18,5])
plt.imshow(cropped_face1); plt.title("Blurred Face");
```
